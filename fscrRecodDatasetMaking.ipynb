```python
import os
import numpy as np
import cv2

data_path = 'train_data_2'

labels = os.listdir(data_path)
print(labels)
categories = np.arange(len(labels))
print(categories)
categories_dict = dict(zip(labels,categories))
print(categories_dict)
```
